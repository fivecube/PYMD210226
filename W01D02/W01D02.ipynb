```python
# conditional statements

age = input('Give us your age')
age = int(age)

if age > 18:
    print('allowed')
    print('you are eligible since you are more than 18 years of age')
else:
    print('not allowed')
```

```python
def calculate_percentage(list_of_marks):
    total_marks = sum(list_of_marks)
    number_of_subjects = len(list_of_marks)
    percentage = total_marks/number_of_subjects
    return percentage

marks_of_students = {
    "mohit": {"maths": 89, "english": 70},
    "rohan": {"maths": 67, "english": 99},
    "aditya": {"maths": 89, "english": 95},
    "ajit": {"maths": 98, "english": 92}
}

for student_name, marks_dict in marks_of_students.items():
    print(f'Student = {student_name}')
    percentage = calculate_percentage(marks_dict.values())
    if percentage >= 90:
        print('Grade A')
    elif percentage >= 80:
        print('Grade B')
    elif percentage >= 70:
        print('Grade C')
    else:
        print('Bad Grade')
```

```python
age = 1

while age < 18:
    print(f'You are {age} years old and hence not allowed to drink alcohol!')
    age = age + 2

print(f"You are {age} years old. Congratulations you can buy alcohol but it is not good for health")
```

```python
num = 0

while num < 10:
    num = num + 1
    if num == 6:
        continue
    print('magic number', num)

print('after the while loop')
```

```python
num = 0

while num < 10:
    num = num + 1
    if num == 6:
        break
    print('magic number', num)

print('after the while loop')
```
