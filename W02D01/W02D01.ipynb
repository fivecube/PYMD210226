```python
def calculate_percantage(list_of_marks):
    total_marks = sum(list_of_marks)
    no_of_subjects = len(list_of_marks)
    per = total_marks/no_of_subjects
    return per

print('blah blah')

print('do as part of cal per')

mohit_marks = [3, 40]

rohit_marks = [30, 10]

mohit_per = calculate_percantage(mohit_marks)
print(f"mohit_per = {mohit_per}")

rohit_per = calculate_percantage(rohit_marks)
print(f"rohit_per = {rohit_per}")
```

```python
calculate_percantage([50, 60])
```

```python
numbers = [1, 2, 3, 4, 5]

# filter(fun, iterable)
# filter returns filtered iterable but it is not readable and hence needs to be
# type casted into something like list

filtered_result = filter(lambda x: x%2==0, numbers)

print(list(filtered_result))
```

```python
magic_list = [12, 23.45, 'magic', 'orange', 'banana', 300]

magic_list[2] = 'abra cadabra'

print(magic_list)
```

```python
# append : add things at the end of the list

magic_list = [12, 23.45, 'magic', 'orange', 'banana', 300]

new_list = []

for item in magic_list:
    if type(item) == type('dummy'):
        new_list.append(item)
```

```python
new_list
```

```python
# extend : adding the items of another list into another list.

magic_list = [12, 23.45, 'magic', 'orange', 'banana', 300]

new_list = ['abra cadabra', 777]

new_list.extend(magic_list)

print(new_list)
```
