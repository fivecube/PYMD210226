```python
marks_of_students = {
    "mohit": {"maths": 89, "english": 70},
    "rohan": {"maths": 67, "english": 99},
    "aditya": {"maths": 89, "english": 95},
    "ajit": {"maths": 98, "english": 92}
}

for key, value in marks_of_students.items():
    print(f"Lets look at data of {key}")
    for key2, value2 in marks_of_students[key].items():
        print(key2, '-->', value2)
```
